```python
# Importar módulos necesarios
import json
import numpy as np
import pickle
```

```python
# Definir función para cargar datos desde JSON
def cargar_datos_desde_json(ruta_archivo):
    with open(ruta_archivo, 'r') as archivo:
        datos = json.load(archivo)
    return datos
```

```python
# Leer datos desde el archivo JSON
ruta_archivo_json = 'sparsification_results.json'  
datos = cargar_datos_desde_json(ruta_archivo_json)
```

## Llaves consideradas:
**lista con un tamaño igual al número de épsilons considerados**

- **epsilon**: Este es el nivel en el que modificamos los elementos de la matriz A. Si el elemento es menor que este nivel (A normalizado), toma el valor 0.
- **objective_function**: Con la nueva matriz A (modificada usando épsilon), construimos un nuevo modelo. Este es el valor de la función objetivo de este nuevo modelo.
- **decision_variables**: Con la nueva matriz A (modificada usando épsilon), construimos un nuevo modelo. Este es el valor de cada variable de decisión para cada épsilon.
- **changed_indices**: Son los índices de la matriz A que fueron modificados considerando el valor épsilon.
- **constraint_violation**: Tomamos la solución del nuevo modelo (matriz A modificada) y la colocamos en el modelo original para evaluar violaciones en cada una de las restricciones.
- **of_original_decision**: Tomamos la solución del nuevo modelo (matriz A modificada) y la colocamos en el modelo original para evaluar el valor de la función objetivo.
